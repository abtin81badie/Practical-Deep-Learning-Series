````python
class MyModel(nn.Module):
    def __init__(self, input_dim, hidden_dims, output_dim):
        super().__init__()
        self.flatten = nn.Flatten()
        self.layers = list()
        self.layers.append(nn.Linear(input_size,hidden_dims[0]))
        self.layers.append(nn.ReLU())
        for idx in range(len(hidden_dims)-1):
            self.layers.append(nn.Linear(hidden_dims[idx],hidden_size[idx+1]))
            self.layers.append(nn.ReLU())
        self.layers.append(nn.Linear(hidden_dims[-1],output_dim))

    def forward(self, X):
        x = self.flatten(X)
        for layer in self.layers:
            x = layer(x)

        return x
````

````python
def accuracy(y_pred, y):
    """
    Calculating the accuracy of given batch.
    input(s):
    y_pred (ndarray): predicted logits or the output of the model.
    y (ndarray): labels
    """
    pred_idx = y_pred.max(1, keepdim=True)[1]
    correct = pred_idx.eq(y.view_as(pred_idx)).sum().item()
    return correct / len(y)
````

<h1>Loading Data</h1>

````python
# 2. Loading Data
def load_data(cancer_path, liver_path):
    # Load Cancer Data
    cancer_x_train = pd.read_csv(f'{cancer_path}/cancer_x_resampled.csv')
    cancer_y_train = pd.read_csv(f'{cancer_path}/cancer_y_resampled.csv')
    cancer_x_test = pd.read_csv(f'{cancer_path}/cancer_x_test.csv')
    cancer_y_test = pd.read_csv(f'{cancer_path}/cancer_y_test.csv')
    
    # Load Liver Data
    liver_x_train = pd.read_csv(f'{liver_path}/liver_x_train.csv')
    liver_y_train = pd.read_csv(f'{liver_path}/liver_y_train.csv')
    liver_x_test = pd.read_csv(f'{liver_path}/liver_x_test.csv')
    liver_y_test = pd.read_csv(f'{liver_path}/liver_y_test.csv')
    
    return {
        'Cancer': {
            'X_train': cancer_x_train,
            'y_train': cancer_y_train,
            'X_test': cancer_x_test,
            'y_test': cancer_y_test
        },
        'Liver': {
            'X_train': liver_x_train,
            'y_train': liver_y_train,
            'X_test': liver_x_test,
            'y_test': liver_y_test
        }
    }
````

<h1>Splitting Data into Training and Validation Sets</h1>

````python
# 3. Splitting Data into Training and Validation Sets
def split_train_val(datasets, val_size=0.1, random_state=42):
    for dataset_name, data in datasets.items():
        X_train = data['X_train']
        y_train = data['y_train']
        
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, 
            test_size=val_size, 
            random_state=random_state, 
            stratify=y_train
        )
        
        datasets[dataset_name]['X_train'] = X_train
        datasets[dataset_name]['y_train'] = y_train
        datasets[dataset_name]['X_val'] = X_val
        datasets[dataset_name]['y_val'] = y_val
    return datasets
````

<h1>Preparing Data for PyTorch</h1>

````python
# 4. Preparing Data for PyTorch
class MedicalDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X.values, dtype=torch.float32)
        self.y = torch.tensor(y.values, dtype=torch.long).squeeze()
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

def create_dataloaders(X_train, y_train, X_val, y_val, X_test, y_test, batch_size=64):
    train_dataset = MedicalDataset(X_train, y_train)
    val_dataset = MedicalDataset(X_val, y_val)
    test_dataset = MedicalDataset(X_test, y_test)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    return train_loader, val_loader, test_loader
````

<h1>Training the Model with Loss adn Accuracy Tracking<h1>

````python
# 6. Training the Model with Loss and Accuracy Tracking
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=50, device='cpu'):
    model.to(device)
    best_val_loss = np.inf
    patience = 5
    trigger_times = 0

    # Lists to store loss and accuracy
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        # Training Phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

            # Calculate accuracy
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        epoch_train_loss = running_loss / len(train_loader.dataset)
        epoch_train_acc = correct / total
        train_losses.append(epoch_train_loss)
        train_accuracies.append(epoch_train_acc)

        # Validation Phase
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)

                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)

        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = correct / total
        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')

        # Early Stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            trigger_times = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

    # Load the best model
    model.load_state_dict(torch.load('best_model.pth'))

    return model, train_losses, val_losses, train_accuracies, val_accuracies
````

<h1>Evaluation the Model and Plotting Results</h1>

````python
# 7. Evaluating the Model and Plotting Results
def evaluate_model(model, test_loader, device='cpu'):
    model.eval()
    model.to(device)
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    # Calculate Accuracy
    acc = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds)
    cm = confusion_matrix(all_labels, all_preds)
    
    print(f'Test Accuracy: {acc:.4f}')
    print('Classification Report:')
    print(report)
    
    # Plot Confusion Matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Negative', 'Positive'], 
                yticklabels=['Negative', 'Positive'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
````

```python
# 5. Defining the MLP Model
class MLP(nn.Module):
    def __init__(self, input_size, hidden_sizes=[128, 64], num_classes=2, dropout_prob=0.5):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_sizes[0])
        self.dropout1 = nn.Dropout(dropout_prob)
        self.fc2 = nn.Linear(hidden_sizes[0], hidden_sizes[1])
        self.dropout2 = nn.Dropout(dropout_prob)
        self.output = nn.Linear(hidden_sizes[1], num_classes)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = self.output(x)
        return x
```

### **Explanation:**

- **`MLP`**: A neural network model with two hidden layers and dropout layers to prevent overfitting.
  - **Parameters**:
    - `input_size`: Number of input features.
    - `hidden_sizes`: List containing the number of neurons in each hidden layer (default is [128, 64]).
    - `num_classes`: Number of output classes for classification (default is 2 for binary classification).
    - `dropout_prob`: Probability of dropping out neurons during training (default is 50%).
  - **Layers**:
    - `fc1`: First fully connected layer.
    - `dropout1`: Dropout layer after the first fully connected layer.
    - `fc2`: Second fully connected layer.
    - `dropout2`: Dropout layer after the second fully connected layer.
    - `output`: Output layer producing logits for each class.
  - **`forward` Method**:
    - Passes input through `fc1` with ReLU activation.
    - Applies `dropout1`.
    - Passes the result through `fc2` with ReLU activation.
    - Applies `dropout2`.
    - Passes the result through the `output` layer to obtain final logits.

## **6. Training the Model with Loss and Accuracy Tracking**

Define a function to train the model, track loss and accuracy for both training and validation sets, and implement early stopping to prevent overfitting.

```python
# 6. Training the Model with Loss and Accuracy Tracking
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=50, device='cpu'):
    model.to(device)
    best_val_loss = np.inf
    patience = 5
    trigger_times = 0

    # Lists to store loss and accuracy
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        # Training Phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

            # Calculate accuracy
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        epoch_train_loss = running_loss / len(train_loader.dataset)
        epoch_train_acc = correct / total
        train_losses.append(epoch_train_loss)
        train_accuracies.append(epoch_train_acc)

        # Validation Phase
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)

                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)

        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = correct / total
        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')

        # Early Stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            trigger_times = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

    # Load the best model
    model.load_state_dict(torch.load('best_model.pth'))

    return model, train_losses, val_losses, train_accuracies, val_accuracies
```

### **Explanation:**

- **Parameters**:
  - `model`: The neural network model to be trained.
  - `train_loader`: DataLoader for the training data.
  - `val_loader`: DataLoader for the validation data.
  - `criterion`: Loss function (e.g., `CrossEntropyLoss`).
  - `optimizer`: Optimization algorithm (e.g., `Adam`).
  - `num_epochs`: Maximum number of training epochs (default is 50).
  - `device`: Computation device (`cpu` or `cuda`).
  
- **Variables**:
  - `best_val_loss`: Tracks the best (lowest) validation loss achieved.
  - `patience`: Number of epochs to wait for improvement before stopping (default is 5).
  - `trigger_times`: Counts the number of consecutive epochs without improvement.
  
- **Process**:
  1. **Training Phase**:
     - Set the model to training mode (`model.train()`).
     - Iterate over batches of training data.
     - Perform forward pass, compute loss, perform backward pass, and update weights.
     - Accumulate loss and correct predictions to compute epoch loss and accuracy.
  
  2. **Validation Phase**:
     - Set the model to evaluation mode (`model.eval()`).
     - Disable gradient computation with `torch.no_grad()`.
     - Iterate over batches of validation data.
     - Perform forward pass and compute loss.
     - Accumulate loss and correct predictions to compute epoch validation loss and accuracy.
  
  3. **Logging**:
     - Print the training and validation loss and accuracy for each epoch.
  
  4. **Early Stopping**:
     - If the validation loss improves, save the model and reset `trigger_times`.
     - If no improvement is observed for `patience` consecutive epochs, stop training early.
  
  5. **Loading the Best Model**:
     - After training, load the model state with the lowest validation loss.

- **Returns**:
  - The trained model.
  - Lists containing training and validation losses and accuracies for each epoch.

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=50, device='cpu'):
```

````python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Loading Data
def load_data(cancer_path, liver_path):
    # Load Cancer Data
    cancer_x_train = pd.read_csv(f'{cancer_path}/cancer_x_train.csv')
    cancer_y_train = pd.read_csv(f'{cancer_path}/cancer_y_train.csv')
    cancer_x_test = pd.read_csv(f'{cancer_path}/cancer_x_test.csv')
    cancer_y_test = pd.read_csv(f'{cancer_path}/cancer_y_test.csv')
    
    # Load Liver Data
    liver_x_train = pd.read_csv(f'{liver_path}/liver_x_train.csv')
    liver_y_train = pd.read_csv(f'{liver_path}/liver_y_train.csv')
    liver_x_test = pd.read_csv(f'{liver_path}/liver_x_test.csv')
    liver_y_test = pd.read_csv(f'{liver_path}/liver_y_test.csv')
    
    return {
        'Cancer': {
            'X_train': cancer_x_train,
            'y_train': cancer_y_train,
            'X_test': cancer_x_test,
            'y_test': cancer_y_test
        },
        'Liver': {
            'X_train': liver_x_train,
            'y_train': liver_y_train,
            'X_test': liver_x_test,
            'y_test': liver_y_test
        }
    }

# 2. Splitting Data into Training and Validation Sets
def split_train_val(datasets, val_size=0.1, random_state=42):
    for dataset_name, data in datasets.items():
        X_train = data['X_train']
        y_train = data['y_train']
        
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, 
            test_size=val_size, 
            random_state=random_state, 
            stratify=y_train
        )
        
        datasets[dataset_name]['X_train'] = X_train
        datasets[dataset_name]['y_train'] = y_train
        datasets[dataset_name]['X_val'] = X_val
        datasets[dataset_name]['y_val'] = y_val
    return datasets

# 3. Preparing Data for PyTorch
class MedicalDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X.values, dtype=torch.float32)
        self.y = torch.tensor(y.values, dtype=torch.long).squeeze()
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

def create_dataloaders(X_train, y_train, X_val, y_val, X_test, y_test, batch_size=64):
    train_dataset = MedicalDataset(X_train, y_train)
    val_dataset = MedicalDataset(X_val, y_val)
    test_dataset = MedicalDataset(X_test, y_test)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    return train_loader, val_loader, test_loader

# 4. Defining the MLP Model
class MLP(nn.Module):
    def __init__(self, input_size, hidden_sizes=[128, 64], num_classes=2, dropout_prob=0.5):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_sizes[0])
        self.dropout1 = nn.Dropout(dropout_prob)
        self.fc2 = nn.Linear(hidden_sizes[0], hidden_sizes[1])
        self.dropout2 = nn.Dropout(dropout_prob)
        self.output = nn.Linear(hidden_sizes[1], num_classes)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = self.output(x)
        return x

# 5. Training the Model with Loss and Accuracy Tracking
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=50, device='cpu'):
    model.to(device)
    best_val_loss = np.inf
    patience = 5
    trigger_times = 0

    # Lists to store loss and accuracy
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        # Training Phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

            # Calculate accuracy
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        epoch_train_loss = running_loss / len(train_loader.dataset)
        epoch_train_acc = correct / total
        train_losses.append(epoch_train_loss)
        train_accuracies.append(epoch_train_acc)

        # Validation Phase
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)

                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)

        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = correct / total
        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')

        # Early Stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            trigger_times = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

    # Load the best model
    model.load_state_dict(torch.load('best_model.pth'))

    return model, train_losses, val_losses, train_accuracies, val_accuracies

# 6. Evaluating the Model and Plotting Results
def evaluate_model(model, test_loader, device='cpu'):
    model.eval()
    model.to(device)
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    # Calculate Accuracy
    acc = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds)
    cm = confusion_matrix(all_labels, all_preds)
    
    print(f'Test Accuracy: {acc:.4f}')
    print('Classification Report:')
    print(report)
    
    # Plot Confusion Matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Negative', 'Positive'], 
                yticklabels=['Negative', 'Positive'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

# 7. Running the Complete Process for Both Datasets
def process_dataset(datasets, dataset_name, num_classes=2, epochs=100):
    print(f'\nProcessing {dataset_name} Dataset...')
    
    data = datasets[dataset_name]
    
    # Create DataLoaders
    train_loader, val_loader, test_loader = create_dataloaders(
        data['X_train'], data['y_train'], 
        data['X_val'], data['y_val'], 
        data['X_test'], data['y_test'], 
        batch_size=64
    )
    
    # Determine input size based on number of features
    input_size = data['X_train'].shape[1]
    
    # Initialize the model
    model = MLP(input_size=input_size, hidden_sizes=[128, 64], num_classes=num_classes, dropout_prob=0.3)
    
    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # Train the model and get loss and accuracy lists
    model, train_losses, val_losses, train_accuracies, val_accuracies = train_model(
        model, train_loader, val_loader, criterion, optimizer, num_epochs=epochs, device=device
    )
    
    # Evaluate the model on the test set
    print(f'\nEvaluating {dataset_name} Model on Test Data:')
    evaluate_model(model, test_loader, device=device)
    
    # Plot Loss
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(train_losses)+1), train_losses, marker='o', label='Training Loss')
    plt.plot(range(1, len(val_losses)+1), val_losses, marker='o', label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{dataset_name} Model Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Plot Accuracy
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(train_accuracies)+1), train_accuracies, marker='o', label='Training Accuracy')
    plt.plot(range(1, len(val_accuracies)+1), val_accuracies, marker='o', label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{dataset_name} Model Training and Validation Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

# 8. Complete Workflow Execution
# Determine the computation device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Define the paths to your datasets
cancer_path = './DataSet/Cancer'
liver_path = './DataSet/Liver'

# Load the data
datasets = load_data(cancer_path, liver_path)

# Split the data into training and validation sets
datasets = split_train_val(datasets, val_size=0.1, random_state=42)

# Function to display dataset information
def display_data_info(datasets):
    for dataset_name, data in datasets.items():
        print(f"\n{dataset_name} Dataset:")
        for split, df in data.items():
            print(f"{split}:")
            print(f"Shape: {df.shape}")
            print("Columns:", df.columns.tolist())
            print()

# Display the dataset information
display_data_info(datasets)

# Process and train the model for each dataset
process_dataset(datasets, 'Cancer', num_classes=2, epochs=100)
process_dataset(datasets, 'Liver', num_classes=2, epochs=100)
````

````python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
from torch.nn.functional import relu

# 1. Define the PyTorch Dataset
class MedicalDataset(Dataset):
    def __init__(self, X, y):
        """
        Args:
            X (pd.DataFrame): Feature data.
            y (pd.Series or np.array): Target labels.
        """
        self.X = torch.tensor(X.values, dtype=torch.float32)
        self.y = torch.tensor(y.values, dtype=torch.long).squeeze()
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# 2. Define the MLP Model
class MLP(nn.Module):
    def __init__(self, input_size, hidden_sizes=[64, 32], num_classes=2, dropout_prob=0.5):
        """
        Args:
            input_size (int): Number of input features.
            hidden_sizes (list): List containing the number of neurons in each hidden layer.
            num_classes (int): Number of output classes.
            dropout_prob (float): Dropout probability.
        """
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_sizes[0])
        self.bn1 = nn.BatchNorm1d(hidden_sizes[0])
        self.dropout1 = nn.Dropout(dropout_prob)
        
        self.fc2 = nn.Linear(hidden_sizes[0], hidden_sizes[1])
        self.bn2 = nn.BatchNorm1d(hidden_sizes[1])
        self.dropout2 = nn.Dropout(dropout_prob)
        
        self.output = nn.Linear(hidden_sizes[1], num_classes)
        
    def forward(self, x):
        x = relu(self.bn1(self.fc1(x)))
        x = self.dropout1(x)
        x = relu(self.bn2(self.fc2(x)))
        x = self.dropout2(x)
        x = self.output(x)
        return x

# 3. Function to Plot Confusion Matrix
def plot_confusion_matrix(cm, classes, title='Confusion Matrix'):
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title(title)
    plt.show()

# 4. Training Function
def train_model(model, train_loader, val_loader, criterion, optimizer, device, num_epochs=100, patience=10):
    """
    Trains the model and implements early stopping.
    
    Args:
        model (nn.Module): The neural network model.
        train_loader (DataLoader): DataLoader for training data.
        val_loader (DataLoader): DataLoader for validation data.
        criterion: Loss function.
        optimizer: Optimizer.
        device: Device to train on ('cpu' or 'cuda').
        num_epochs (int): Maximum number of epochs.
        patience (int): Number of epochs to wait for improvement before stopping.
        
    Returns:
        model: Trained model with best validation loss.
        history: Dictionary containing training and validation loss and accuracy.
    """
    model.to(device)
    best_val_loss = np.inf
    trigger_times = 0
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    for epoch in range(num_epochs):
        # Training Phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0
        
        for inputs, labels in train_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        epoch_train_loss = running_loss / total
        epoch_train_acc = correct / total
        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc)
        
        # Validation Phase
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                
                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)
        
        epoch_val_loss = val_loss / total
        epoch_val_acc = correct / total
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)
        
        print(f'Epoch {epoch+1}/{num_epochs}: '
              f'Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}')
        
        # Early Stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            trigger_times = 0
            best_model_wts = model.state_dict()
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print(f'Early stopping triggered at epoch {epoch+1}')
                break
    
    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model, history

# 5. Evaluation Function
def evaluate_model_pytorch(model, test_loader, device, classes=['Negative', 'Positive']):
    """
    Evaluates the model on the test set and prints metrics.
    
    Args:
        model (nn.Module): Trained model.
        test_loader (DataLoader): DataLoader for test data.
        device: Device to perform evaluation on.
        classes (list): List of class names.
    """
    model.eval()
    model.to(device)
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    acc = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=classes, digits=4)
    cm = confusion_matrix(all_labels, all_preds)
    
    print(f'\nTest Accuracy: {acc:.4f}')
    print('Classification Report:')
    print(report)
    plot_confusion_matrix(cm, classes, title='Confusion Matrix')

# 6. Function to Create DataLoaders
def create_dataloaders(X_train, y_train, X_val, y_val, X_test, y_test, batch_size=32):
    """
    Creates DataLoaders for training, validation, and testing.
    
    Args:
        X_train, y_train: Training data and labels.
        X_val, y_val: Validation data and labels.
        X_test, y_test: Test data and labels.
        batch_size (int): Batch size.
        
    Returns:
        train_loader, val_loader, test_loader: DataLoaders for each dataset split.
    """
    train_dataset = MedicalDataset(X_train, y_train)
    val_dataset = MedicalDataset(X_val, y_val)
    test_dataset = MedicalDataset(X_test, y_test)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    return train_loader, val_loader, test_loader

# 7. Function to Display Dataset Information with Type Checking
def display_dataset_info(datasets):
    print("\n=== Dataset Information ===")
    for dataset_name, data in datasets.items():
        print(f"\n{dataset_name} Dataset:")
        for split, df in data.items():
            if isinstance(df, pd.DataFrame):
                print(f"{split}: Shape {df.shape}, Columns: {df.columns.tolist()}")
            else:
                print(f"{split}: Not a DataFrame, type: {type(df)}")

# 8. Main Function to Process Datasets
def main():
    from sklearn.model_selection import train_test_split
    
    # Define the paths to your datasets
    cancer_path = './DataSet/CancerResampled'
    liver_path = './DataSet/LiverResampled'
    
    # Load the data
    print("\nLoading data...")
    datasets = load_data(cancer_path, liver_path)
    
    # Display Dataset Information
    display_dataset_info(datasets)
    
    # Preprocess and scale the data
    datasets = preprocess_data(datasets)
    
    # Split training data into training and validation sets, then perform feature selection
    for dataset_name in ['Cancer', 'Liver']:
        print(f"\n--- Splitting {dataset_name} Dataset into Train and Validation ---")
        data = datasets[dataset_name]
        X_train_full = data['X_train_scaled']
        y_train_full = data['y_train_resampled']
        
        X_train_final, X_val, y_train_final, y_val = train_test_split(
            X_train_full, y_train_full,
            test_size=0.2,
            random_state=42,
            stratify=y_train_full
        )
        
        datasets[dataset_name]['X_train_final'] = X_train_final
        datasets[dataset_name]['y_train_final'] = y_train_final
        datasets[dataset_name]['X_val'] = X_val
        datasets[dataset_name]['y_val'] = y_val
    
    # Perform Feature Selection on the training set
    for dataset_name in ['Cancer', 'Liver']:
        print(f"\n--- Feature Selection for {dataset_name} Dataset ---")
        data = datasets[dataset_name]
        selector = SelectKBest(score_func=f_classif, k=20)
        selector.fit(data['X_train_final'], data['y_train_final'])
        
        X_train_selected = selector.transform(data['X_train_final'])
        X_val_selected = selector.transform(data['X_val'])
        X_test_selected = selector.transform(data['X_test_scaled'])
        
        selected_feature_names = data['X_train_final'].columns[selector.get_support()]
        print(f"Selected Features ({len(selected_feature_names)}): {selected_feature_names.tolist()}")
        
        # Update datasets with selected features
        datasets[dataset_name]['X_train_selected'] = pd.DataFrame(X_train_selected, columns=selected_feature_names)
        datasets[dataset_name]['X_val_selected'] = pd.DataFrame(X_val_selected, columns=selected_feature_names)
        datasets[dataset_name]['X_test_selected'] = pd.DataFrame(X_test_selected, columns=selected_feature_names)
    
    # Now, create DataLoaders and train the MLP model for each dataset
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'\nUsing device: {device}')
    
    for dataset_name in ['Cancer', 'Liver']:
        print(f'\n--- Processing {dataset_name} Dataset with MLP ---')
        
        data = datasets[dataset_name]
        X_train = data['X_train_selected']
        y_train = data['y_train_final']
        X_val = data['X_val_selected']
        y_val = data['y_val']
        X_test = data['X_test_selected']
        y_test = data['y_test']
        
        # Create DataLoaders
        train_loader, val_loader, test_loader = create_dataloaders(
            X_train, y_train, X_val, y_val, X_test, y_test, batch_size=32
        )
        
        # Determine input size
        input_size = X_train.shape[1]
        
        # Initialize the MLP model
        model = MLP(input_size=input_size, hidden_sizes=[64, 32], num_classes=2, dropout_prob=0.5)
        
        # Compute class weights to handle class imbalance
        classes = np.unique(y_train)
        class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
        class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
        print(f"\nClass Weights for {dataset_name}: {class_weights.cpu().numpy()}")
        
        # Define loss function with class weights
        criterion = nn.CrossEntropyLoss(weight=class_weights)
        
        # Define optimizer with weight decay for regularization
        optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
        
        # Train the model
        model, history = train_model(model, train_loader, val_loader, criterion, optimizer, device, num_epochs=100, patience=10)
        
        # Evaluate the model on the test set
        evaluate_model_pytorch(model, test_loader, device, classes=['Negative', 'Positive'])
        
        # Plot Training and Validation Loss
        plt.figure(figsize=(10,5))
        plt.plot(range(1, len(history['train_loss'])+1), history['train_loss'], label='Train Loss')
        plt.plot(range(1, len(history['val_loss'])+1), history['val_loss'], label='Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title(f'{dataset_name} MLP Model Training and Validation Loss')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        # Plot Training and Validation Accuracy
        plt.figure(figsize=(10,5))
        plt.plot(range(1, len(history['train_acc'])+1), history['train_acc'], label='Train Accuracy')
        plt.plot(range(1, len(history['val_acc'])+1), history['val_acc'], label='Validation Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.title(f'{dataset_name} MLP Model Training and Validation Accuracy')
        plt.legend()
        plt.grid(True)
        plt.show()

# 9. Execute the Main Function
if __name__ == "__main__":
    main()
````
